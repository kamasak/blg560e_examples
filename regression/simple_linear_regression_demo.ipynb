It is generally accepted that by increasing the number of produced units, cost per unit can be decreased linearly. A manufacturer records the number of units and cost per unit as follows:

|# of units| Cost|
|:--------:|:--:|
| 10 |9.4|
| 20 | 9.2|
| 50 | 9.0|
| 100 |8.5|
| 150 |8.1|
|200 |7.4|

a. Predict the cost per unit when 125 units are produced.

```python
import numpy as np
import matplotlib.pyplot as plt

items = np.array([10, 20, 50, 100, 150, 200])
cost = np.array([9.4, 9.2, 9.0, 8.5, 8.1, 7.4])

# Lets visualize the relation between the number of items 
# and cost per unit
plt.scatter(items, cost)
plt.show();
```

```python
# Relation between the number of items and cost per unit
#   seems linear
# Hence simple linear regression can be used

# input: number of items
# output: cost per unit

N = items.shape[0]

Sxx = (items**2).sum() - N*items.mean()**2
Sxy = np.multiply(items, cost).sum()-N*items.mean()*cost.mean()
print(f"Sxx: {Sxx:.6} Sxy: {Sxy:.4}")
beta1 = Sxy/Sxx

beta0 = cost.mean()-beta1*items.mean()
print(f"Parameters beta0: {beta0:.4} beta1: {beta1:.4}")

plt.scatter(items, cost)
plt.plot(items, beta0+beta1*items, color='red')
plt.legend(['model', 'observation'])
plt.show();
```

```python
# Using model predict cost per unit for 125 units

unit = 125
cost1 = beta1*unit + beta0 

print(f"Cost per unit for {unit} items: {cost1:.2}")
```

```python
# Lets visualize the model prediction

plt.scatter(items, cost)
plt.plot(items, beta0+beta1*items, color='red')
plt.scatter(unit, cost1, color='green')
plt.legend(['model', 'observation'])
plt.show();
```

b. Estimate the variance of the cost in part (a).

Model variance is:
$$
\sigma^2 = \frac{SSE}{N-2}
$$

```python
cost_model = beta1*items + beta0 
SSE = ((cost-cost_model)**2).sum()
print(f"SSE: {SSE}")
sigma2 = SSE/(N-2)

print(f"Model variance: {sigma2:.4}")
```

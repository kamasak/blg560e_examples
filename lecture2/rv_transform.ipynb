## Random variable transformation 

Let X be a random variable with exponential distribution.

$f_X(x)=\lambda e^{-\lambda x}$ if x>=0, 0 otherwise

$F_X(x)=1-e^{-\lambda x}$

```python
from scipy.stats import expon
import matplotlib.pyplot as plt

# Let lambda=0.2, then variance = 1/lambda^2
x = expon.rvs(scale=5,size=50000)

plt.hist(x,bins=100,density=True);
```

Lets convert $X$ to $Y$ where $Y$ has uniform distribution.

$y = F_X(x)$

```python
import math 

# F(X)=1-e^{-lambda x}
y = list(map(lambda x: 1-math.exp(-0.2*x), x))


plt.hist(y,bins=40,density=True);
```

Random variable $Y$ has uniform distribution. Now, lets convert $Y$ to $Z$ where $Z$ has standard normal distribution.

$z = \Phi^{-1}(y)$ where $\Phi$ is the cdf of standard normal distribution.

```python
from scipy.stats import norm

z = map(lambda x: norm.ppf(x), y)

plt.hist(list(z),bins=100,density=True);
```

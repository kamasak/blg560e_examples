## Demo for central limit theorem (CLT)

Let $X_i$ for $i=0,1, ..., N-1$ be iid random variable coming from exponential distribution with $\lambda=5$.

$$
f(X_i)=\lambda\exp(-\lambda X_i)
$$

Define $\overline{X} = \frac{1}{N}\sum_i X_i$

Then show $\overline{X} \sim {\cal N}(\frac{1}{\lambda}, \frac{1}{\lambda^2 N})$

```python
from scipy.stats import norm, expon
import numpy as np
import matplotlib.pyplot as plt

# Generate N random numbers from exp distr with lambda=5
# return its mean (x_bar)
def generate_random_variables(N):
    x = expon.rvs(scale=0.2, size=N)
    return np.sum(x)/N

x = generate_random_variables(100)
print(x)
```

Now generate random variables (for N=100) and record their mean for 3000 times. Then plot these mean values' histogram.

```python
N = 100
M = 3000
x = [generate_random_variables(N) for _ in range(M)]

plt.hist(x, bins=50, density=True);
plt.xlabel('X');
```

Overlay ${\cal N}(\frac{1}{\lambda}, \frac{1}{\lambda^2 N})$ on top of the histogram.

```python

r = np.arange(3.4,6.6,0.01)
norm_rv = norm.pdf(r, loc=1/0.2, scale=1/(0.2*np.sqrt(100)))

plt.hist(x, bins=50, density=True)
plt.plot(r, norm_rv, color='red')
plt.legend(['norm distr','mean hist'])
```
